```python
def Division_zone(n,m):
    first_lat = 37.413294
    first_log = 126.734086
    lat_side_length = (37.715133 - 37.413294) / n
    log_side_length = (127.269311 - 126.734086) / m
    DF = pd.DataFrame()
#     DF = pd.DataFrame(columns=["first_lat", "firtst_log", "last_lat", "last_log"])
#     Count = n * m
    
    for i in range(n):
        last_lat = first_lat + lat_side_length
        index = i * m
        for j in range(m):
            last_log = first_log + log_side_length
            DF[index+j] = [first_lat, first_log, last_lat, last_log]
            first_log = last_log
        first_lat = last_lat
        first_log = 126.734086
        
#     for i in range(Count):
#         last_lat = first_lat + lat_side_length
#         last_log = first_log + log_side_length
#         DF[i] = [first_lat, first_log, last_lat, last_log]
#         first_lat = last_lat
#         first_log = last_log
        
    return DF
```
