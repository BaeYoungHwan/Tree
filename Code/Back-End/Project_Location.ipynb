```python
def Tree_separate(first_lat, first_log, last_lat, last_log, treeset):
    DF = pd.DataFrame(columns=treeset.columns)
    a = 0
    for i in treeset.iterrows():
        if (first_lat < i[1][1] < last_lat) & (first_log < i[1][2] < last_log):
            DF.loc[a] = i[1]
            a += 1
    return DF
```
