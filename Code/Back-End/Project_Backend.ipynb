#### Location 위치 잡는 방법
1. 모든 Location을 임의의 크기로 잡는다
2. 건물의 면적 또는 높이에 따라 크기를 다르게 잡는다

-------- Code --------

```python
import numpy as np
```

```python
# 중심 아님
def SelectLocation(lng, lat):
    length = 0.00003
    return np.array([[lng, lat+length], [lng+length, lat]], dtype=np.float64)
```

```python
SelectLocation(Junggu_building_location.loc[0][0], Junggu_building_location.loc[0][1])[0]
```

```python
tree.iloc[0]
```

```python
def AreaCountTrees(lng, lat, dataset):
    area = SelectLocation(lng, lat)
    area_tree = pd.DataFrame(columns=list(set(tree['수목명'])))

    area_tree.loc[0] = (np.zeros(len(area_tree.columns)))

    for i in dataset.iterrows():

        if (area[0][0] < i[1][1] < area[1][0]) & (area[1][1] < i[1][2] < area[0][1]):

            area_tree[i[1][0]] += 1
    
    print(area_tree.sort_values(by=0, ascending=False, axis=1))
    return area_tree
```

```python
pd.concat([AreaCountTrees(37.562691,126.99801,tree),AreaCountTrees(37.5604659, 127.0033242,tree)])
```

```python
%%time
DB = pd.concat([AreaCountTrees(x[1][0], x[1][1], tree) for x in Junggu_building_location.head(1000).iterrows()])
```
